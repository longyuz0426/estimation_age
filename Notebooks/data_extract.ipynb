```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#lecture des fichiers csv
train = pd.read_csv("fairface_label_train.csv")
val = pd.read_csv("fairface_label_val.csv")
```

```python
train.head()
```

```python
#extraction de la colonne contenant le chemin des images
train_files = train['file']
val_files = val['file']
```

```python
#on precise le repertoire : padding_025 o padding 0125
train_files_025 = "padding_025/" + train_files
train_files_125 = "padding_125/" + train_files
val_files_025 = "padding_025/" + val_files
val_files_125 = "padding_125/" + val_files
print("train_025 : ", train_files_025.shape)
print("train_125 : ", train_files_125.shape)
print("val_025 : ", val_files_025.shape)
print("val_125 : ", val_files_125.shape)
```

```python
#images train pour les 2 padding

data_train_025 = []
data_train_125 = []
#on met dans data_train_025 toute les images train de padding_025
#on met dans data_train_125 toute les images train de padding_125

for i in range(20000) : 
    #nom du fichier
    file025 = train_files_025[i]
    #file125 = train_files_125[i]
    
    #lecture de l'image
    img025 = plt.imread(file025)
    #img125 = plt.imread(file125)
    
    #append
    data_train_025.append(img025)
    #data_train_125.append(img125)
    
```

```python
#affichage 3 premieres images train
data_train_025 = np.array(data_train_025)
print("data_train_025 : ",data_train_025.shape)
plt.figure()
plt.imshow(data_train_025[0])
plt.figure()
plt.imshow(data_train_025[1])
plt.figure()
plt.imshow(data_train_025[2])
```

```python
#images val pour les 2 padding

data_val_025 = []
data_val_125 = []
#on met dans data_val_025 toute les images val de padding_025
#on met dans data_val_125 toute les images val de padding_125

for i in range(len(val_files_025)) : 
    #nom du fichier
    file025 = val_files_025[i]
    #file125 = val_files_125[i]
    
    #lecture de l'image
    img025 = plt.imread(file025)
    #img125 = plt.imread(file125)
    
    #append
    data_val_025.append(img025)
    #data_val_125.append(img125)

    
```

```python
#affichage 3 premieres images de val
data_val_025 = np.array(data_val_025)
print("data_val_025 : ",data_val_025.shape)
plt.figure()
plt.imshow(data_val_025[0])
plt.figure()
plt.imshow(data_val_025[1])
plt.figure()
plt.imshow(data_val_025[2])
```
